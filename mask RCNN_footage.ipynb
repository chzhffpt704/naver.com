```python
import os
import sys
import cv2

import numpy as np

#라이브러리를 불러오기위한 path 조정
ROOT_DIR = os.path.abspath("../")
sys.path.append(ROOT_DIR)
COCO_DIR=os.path.join(ROOT_DIR, "samples/coco/")
sys.path.append(COCO_DIR)

from mrcnn import utils  #marcnn 3개 불러오기
import mrcnn.model as modellib
from visualize_cv2 import model, display_instances, class_names



capture = cv2.VideoCapture('C:/Users/Administrator/Desktop/학교/python/cctv.avi')

size = (
    int(capture.get(cv2.CAP_PROP_FRAME_WIDTH)),
    int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
)


codec = cv2.VideoWriter_fourcc(*'MJPG')
# 가능한 코덱 DIVX, XVID, MJPG, X264, WMV1, WMV2 
output = cv2.VideoWriter('avideofile_masked.avi', codec, 4.0, size)
#cv2.VideoWriter(filename,압축 관련 4자리 code, 초당 저장할 frame, 프레임 사이즈, 컬러 저장여부)
```

```python
while(capture.isOpened()):

    ret, frame = capture.read()

    if ret:

        # add mask to frame

        results = model.detect([frame], verbose=0)

        r = results[0]

        frame = display_instances(

            frame, r['rois'], r['masks'], r['class_ids'], class_names, r['scores']

        )
        output.write(frame)
        cv2.imshow('frame', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break


capture.release()
output.release()
cv2.destroyAllWindows()
```
